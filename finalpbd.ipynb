```python
import os 
folderPath=r"E:\wiki"
filePath=[os.path.join(folderPath,name) for name in os.listdir(folderPath) ]
files_array=[]
for path in filePath:
    with open(path,'r',encoding="ISO-8859-1" )as f:
        file=f.readlines()
        files_array.append(file)
```

```python
files_array
```

```python
lst2str=' '.join(map(str,files_array))
lst2str
```

```python
#importing nltk
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize 
```

```python
txt= word_tokenize(lst2str)
```

```python
txt
```

```python
data=[word for word in txt if not word in stopwords.words()]
print(data)
```

```python
nltk.download('wordnet')
```

```python
 nltk.download('omw-1.4')
```

```python
#lemmating data
from nltk.stem import WordNetLemmatizer
lemmmatizer=WordNetLemmatizer()
lem=[]
for x in data:
    lem.append(lemmmatizer.lemmatize(x))

#importing re library
import re
#updating data
updatedData=[re.sub('[^a-zA-Z0-9]+','',_) for _ in lem]
updatedData
```

```python
# now forming clusters of the list lem[]
!pip install scikit-learn
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.cluster import KMeans
from warnings import simplefilter 

simplefilter(action='ignore' , category=FutureWarning)
vectorizer= TfidfVectorizer(stop_words='english')
x=vectorizer.fit_transform(updatedData)

k=6
model=KMeans(n_clusters=k,init='k-means++' ,max_iter=100 ,n_init=1)
model.fit(x)

print("Top per cluster ")
order_centroids=model.cluster_centers_.argsort()[:, ::-1] 
terms =vectorizer.get_feature_names()
for i in range(k):
    print("Cluster %d :" % i), 
    for ind in order_centroids[i, :10]:
        print('%s' % terms[ind])
    
                                        
```

```python
print('clustters are: \n')


Y = vectorizer.transform(["family"]) 
guessing = model.predict(Y) 
print(guessing)

```
